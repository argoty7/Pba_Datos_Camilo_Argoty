```python
# Carga de archivo de Excel
df = pd.read_excel('articles-391559_recurso.xlsx')
```

```python
# Detección del número de filas a omitir
n = 0
for index, row in df.iterrows():
    if row.notnull().all():  # Si todas las celdas de la fila no están vacías
        break
    n += 1
```

```python
# Omitir las filas no deseadas
df = df.iloc[n:]
```

```python
df.head(10)
```

```python
# Eliminación los caracteres \n en los valores del DataFrame
df = df.replace(r'\n', ' ', regex=True)
```

```python
# Convertir la primera fila en encabezado
df.columns = df.iloc[0]
df = df[1:]
```

```python
# Verificación encabezados
df.head(10)
```

```python
# Eliminación de valores nulos o faltantes
df = df.dropna()
```

```python
# Importación de método unicodedata para eliminar tildes
import unicodedata
```

```python
# Normalización NFKD de encabezados para eliminar tíldes y demás caracteres no ASCII
df.columns = [unicodedata.normalize('NFKD', header).encode('ASCII', 'ignore').decode('utf-8') for header in df.columns]
```

```python
# Normalización NFKD de valores para eliminar tíldes y demás caracteres no ASCII
df = df.applymap(lambda x: unicodedata.normalize('NFKD', x).encode('ASCII', 'ignore').decode('utf-8') if isinstance(x, str) else x)
```

```python
# Conversión a letras mayúsculas de los encabezados
df.columns = [column.upper() for column in df.columns]
```

```python
# Conversión de los valores a letras mayúsculas
df = df.applymap(lambda x: x.upper() if isinstance(x, str) else x)
```

```python
# Verificación conversión a mayúsculas y eliminación de caracteres no ASCII
df.sample(10)
```

```python
# Importación de SpaCy
import spacy
```

```python
# Carga del modelo de idioma en español de spaCy
nlp = spacy.load('es_core_news_sm')

# Función para lematizar un texto
def lemmatize_text(text):
    if isinstance(text, str):
        doc = nlp(text)
        lemmatized_text = " ".join([token.lemma_ for token in doc])
        return lemmatized_text
    return text
```

```python
# Verificación de los nombres de las columnas para selección y reducción de redundancia
df.columns
```

```python
#for column in df.columns:
df['DEPARTAMENTO DE \nDOMICILIO DE LA IES'] = df['DEPARTAMENTO DE \nDOMICILIO DE LA IES'].apply(lemmatize_text)
```

```python
df['MUNICIPIO DE\nDOMICILIO DE LA IES'] = df['MUNICIPIO DE\nDOMICILIO DE LA IES'].apply(lemmatize_text)
```

```python
df['NUCLEO BASICO DEL CONOCIMIENTO (NBC)'] = df['NUCLEO BASICO DEL CONOCIMIENTO (NBC)'].apply(lemmatize_text)
```

```python
# Reducir redundancia en los datos utilizando lematización para todas las columnas
for column in df.columns:
    df[column] = df[column].apply(lemmatize_text)
```

```python
# Verificación conversión a mayúsculas y eliminación de caracteres no ASCII
df.sample(10)
```

```python
# Guardado de los datos resultantes en un archivo para 
df.to_excel('datos_procesados.xlsx', index=False)
```
