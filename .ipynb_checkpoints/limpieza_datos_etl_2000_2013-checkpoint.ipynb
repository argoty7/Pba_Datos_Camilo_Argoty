```python
import pandas as pd
import pyxlsb
```

```python
# Carga de archivo de Excel
df = pd.read_excel('articles-391562_recurso.xlsb')
```

```python
# Detección del número de filas a omitir
n = 0
for index, row in df.iterrows():
    if row.notnull().all():  # Si todas las celdas de la fila no están vacías
        break
    n += 1
```

```python
# Omitir las filas no deseadas
df = df.iloc[n:]
```

```python
# Eliminación los caracteres \n en los valores del DataFrame
df = df.replace(r'\n', ' ', regex=True)
```

```python
# Importación de SpaCy
import spacy
```

```python
# Carga del modelo de idioma en español de spaCy
nlp = spacy.load('es_core_news_sm')

# Función para lematizar un texto
def lemmatize_text(text):
    if isinstance(text, str):
        doc = nlp(text)
        lemmatized_text = " ".join([token.lemma_ for token in doc])
        return lemmatized_text
    return text
```

```python
# Verificación de los nombres de las columnas para selección y reducción de redundancia
df.columns
```

```python
#for column in df.columns:
df['DEPARTAMENTO DE \nDOMICILIO DE LA IES'] = df['DEPARTAMENTO DE \nDOMICILIO DE LA IES'].apply(lemmatize_text)
```

```python
df['MUNICIPIO DE\nDOMICILIO DE LA IES'] = df['MUNICIPIO DE\nDOMICILIO DE LA IES'].apply(lemmatize_text)
```

```python
df['NUCLEO BASICO DEL CONOCIMIENTO (NBC)'] = df['NUCLEO BASICO DEL CONOCIMIENTO (NBC)'].apply(lemmatize_text)
```

```python
# Reducir redundancia en los datos utilizando lematización para todas las columnas
for column in df.columns:
    df[column] = df[column].apply(lemmatize_text)
```

```python
# Verificación conversión a mayúsculas y eliminación de caracteres no ASCII
df.sample(10)
```
