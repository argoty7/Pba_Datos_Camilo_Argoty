```python
import pandas as pd
import numpy as np
# Importación de método unicodedata para eliminar tildes
import unicodedata

# Importación de librería de procesamiento de lenguaje natural nltk
import nltk
from nltk.corpus import stopwords

# Importar librería re para expresiones regulares
import re

# Importar librería pyxlsb para archivos binarios de Excel
import pyxlsb
```

```python
# Descargar las palabras de parada en español (solo se necesita hacer una vez)
nltk.download('stopwords')

# Obtener el conjunto de palabras de parada en español
stopwords_es = set(stopwords.words('spanish'))
```

```python
# Función para filtrar palabras frecuentes de un encabezado
def filtrar_palabras(encabezado):
    palabras = re.findall(r'\b\w+\b', encabezado)  # Obtener todas las palabras en el encabezado
    palabras_filtradas = [p for p in palabras if p.lower() not in stopwords_es]  # Filtrar palabras frecuentes
    return ' '.join(palabras_filtradas)  # Unir las palabras filtradas en un solo encabezado
```

```python
# Lista de nombres de archivos de Excel
archivos_excel = ['articles-391557_recurso.xlsx', 'articles-391558_recurso.xlsx', 'articles-391559_recurso.xlsx', 'articles-391560_recurso.xlsx', 'articles-391561_recurso.xlsx','articles-391562_recurso.xlsb', 'articles-401907_recurso.xlsx', 'articles-406670_recurso.xlsx', 'articles-411251_recurso.xlsx']
```

```python
# Lista para almacenar los DataFrames de los archivos procesados
dataframes_cargados = []

for archivo in archivos_excel:
    
    # Carga de archivo de Excel
    df = pd.read_excel(archivo)
    
    # Detección del número de filas a omitir
    n = 0
    for index, row in df.iterrows():
        if row.notnull().all():  # Si todas las celdas de la fila no están vacías
            break
        n += 1
    
    # Omitir las filas no deseadas
    df = df.iloc[n:]
    
    # Eliminación los caracteres \n en los valores del DataFrame
    df = df.replace(r'\n', ' ', regex=True)
    
    # Convertir la primera fila en encabezado
    df.columns = df.iloc[0]
    df = df[1:]
    
    # Eliminación de valores nulos o faltantes
    df = df.dropna()
    # Agregar el DataFrame procesado a la lista
    dataframes_cargados.append(df)
```

```python
i = 0

dataframes_procesados = []

for df in dataframes_cargados:
    # Transformar el archivo de 2000-2013
    if i==5:
                
        # Definición de las columnas a transformar
        columnas_a_convertir = ['Hombre 2000-1','Mujer 2000-1','Total 2000-1','Hombre 2000-2','Mujer 2000-2','Total 2000-2','Hombre 2001-1','Mujer 2001-1','Total 2001-1','Hombre 2001-2','Mujer 2001-2','Total 2001-2','Hombre 2002-1','Mujer 2002-1','Total 2002-1','Hombre 2002-2','Mujer 2002-2','Total 2002-2','Hombre 2003-1','Mujer 2003-1','Total 2003-1','Hombre 2003-2','Mujer 2003-2','Total 2003-2','Hombre 2004-1','Mujer 2004-1','Total 2004-1','Hombre 2004-2','Mujer 2004-2','Total 2004-2','Hombre 2005-1','Mujer 2005-1','Total 2005-1','Hombre 2005-2','Mujer 2005-2','Total 2005-2','Hombre 2006-1','Mujer 2006-1','Total 2006-1','Hombre 2006-2','Mujer 2006-2','Total 2006-2','Hombre 2007-1','Mujer 2007-1','Total 2007-1','Hombre 2007-2','Mujer 2007-2','Total 2007-2','Hombre 2008-1','Mujer 2008-1','Total 2008-1','Hombre 2008-2','Mujer 2008-2','Total 2008-2','Hombre 2009-1','Mujer 2009-1','Total 2009-1','Hombre 2009-2','Mujer 2009-2','Total 2009-2','Hombre 2010-1','Mujer 2010-1','Total 2010-1','Hombre 2010-2','Mujer 2010-2','Total 2010-2','Hombre 2011-1','Mujer 2011-1','Total 2011-1','Hombre 2011-2','Mujer 2011-2','Total 2011-2','Hombre 2012-1','Mujer 2012-1','Total 2012-1','Hombre 2012-2','Mujer 2012-2','Total 2012-2','Hombre 2013-1*','Mujer 2013-1*','Total 2013-1*','Hombre 2013-2*','Mujer 2013-2*']
        
        # Identificadores de las demás columnas
        id_vars = [col for col in df.columns if col not in columnas_a_convertir]
        
        # Realizar la conversión de columnas a filas
        df = pd.melt(df, id_vars=id_vars, value_vars=columnas_a_convertir, var_name='Categoria', value_name='Valor')
        
        # Eliminar filas con 'Total' en la columna 'Categoria' del DataFrame df
        df = df[~df['Categoria'].str.contains('Total')]
        
        # Separar la columna 'Categoria' en 'Sexo', 'Año' y 'Semestre'
        df[['SEXO', 'ANO', 'SEMESTRE']] = df['Categoria'].str.replace('*', '').str.split(' |-', expand=True)
        
        # Eliminar columnas 'Total 2013-2*' y 'Categoria' del DataFrame df
        df = df.drop(['Total 2013-2*', 'Categoria'], axis=1)
        
        # Cambiar el nombre de la columna 'valor' a 'Inscritos'
        df = df.rename(columns={'Valor': 'Inscritos'})
        
        # Eliminación de valores nulos o faltantes
        df = df.dropna()
        
        # Selección de la columna 'Inscritos' para llevarla al final
        selected_column = 'Inscritos'
        
              
        # Crear lista de columnas
        other_columns = [col for col in df.columns if col != selected_column]

        # Ubicación de la columna 'Inscritos al final
        df = df[other_columns + [selected_column]]
        
    
    # Obtener el nombre de la última columna
    ultima_columna = df.columns[-1]
    
    # Cambiar el nombre de la última columna a 'Inscritos'
    df = df.rename(columns={ultima_columna: 'Inscritos'})
    
    # Normalización NFKD de encabezados para eliminar tíldes y demás caracteres no ASCII
    df.columns = [unicodedata.normalize('NFKD', header).encode('ASCII', 'ignore').decode('utf-8') for header in df.columns]
    
    # Normalización NFKD de valores para eliminar tíldes y demás caracteres no ASCII
    df = df.applymap(lambda x: unicodedata.normalize('NFKD', x).encode('ASCII', 'ignore').decode('utf-8') if isinstance(x, str) else x)
    
    # Conversión a letras mayúsculas de los encabezados
    df.columns = [column.upper() for column in df.columns]
    
    # Conversión de los valores a letras mayúsculas
    df = df.applymap(lambda x: x.upper() if isinstance(x, str) else x)
    
    # Reemplazar los guiones bajos por espacios en los encabezados del DataFrame
    df.columns = df.columns.str.replace('_', ' ')
        
    # Eliminar stopwords de los encabezados del DataFrame
    df.columns = df.columns.to_series().apply(filtrar_palabras)
    
    # Eliminar comas de los valores del DataFrame
    df = df.replace(',', '', regex=True)
    
    # Eliminar palabras repetidas en los encabezados
    df.columns = pd.Index([' '.join(list(dict.fromkeys(col.split()[::-1]))[::-1]) for col in df.columns])
    
    # Cambiar el nombre de la columna 'METODOLOGIA PROGRAMA' a 'Inscritos'
    df = df.rename(columns={'METODOLOGIA PROGRAMA': 'METODOLOGIA'})

    # Verificar si la columna 'GENERO' está presente en el DataFrame
    if 'GENERO' in df.columns:
        # Reemplazar los valores de 'MASCULINO' por 'HOMBRE' y 'FEMENINO' por 'MUJER'
        df.loc[df['GENERO'] == 'MASCULINO', 'GENERO'] = 'HOMBRE'
        df.loc[df['GENERO'] == 'FEMENINO', 'GENERO'] = 'MUJER'
        # Cambiar el nombre de la columna 'GENERO' a 'SEXO'
        df = df.rename(columns={'GENERO': 'SEXO'})
    
    # Eliminar todas las columnas cuyos encabezados empiecen por 'ID'
    columnas_sin_id = [col for col in df.columns if not col.startswith('ID')]
    df = df[columnas_sin_id]
    
    # Filtrar y eliminar los registros que contienen 'NO APLICA', 'N/A' o 'NA' en algún campo
    df = df[~df.isin(['NO APLICA', 'N/A', 'NA','NO IDENTIFICADO','SIN PROGRAMA ESPECIFICO','SIN PROGRAMA ESPECIFICOA','SIN CLASIFICAR']).any(axis=1)]
    
    # Eliminar los registros donde df['INSCRITOS'] == 0
    df = df[df['INSCRITOS'] != 0]
    
    # Modificar los valores de la columna 'MUNICIPIO OFERTA PROGRAMA' si 'DEPARTAMENTO OFERTA PROGRAMA' es 'BOGOTA D.C.'
    df['MUNICIPIO OFERTA PROGRAMA'] = np.where(df['DEPARTAMENTO OFERTA PROGRAMA'] == 'BOGOTA D.C.', 'BOGOTA', df['MUNICIPIO OFERTA PROGRAMA'])
    df['MUNICIPIO OFERTA PROGRAMA'] = np.where(df['DEPARTAMENTO OFERTA PROGRAMA'] == 'BOGOTA D.C', 'BOGOTA', df['MUNICIPIO OFERTA PROGRAMA'])
       
    
    # Agregar el DataFrame procesado a la lista
    dataframes_procesados.append(df)
    i += 1
    
```

```python
# Obtener la intersección de las columnas de todos los DataFrames
columnas_interseccion = set.intersection(*[set(df.columns) for df in dataframes_procesados])
columnas_interseccion
```

```python
# Eliminar las columnas que no están en la intersección de cada DataFrame
for df in dataframes_procesados:
    columnas_a_eliminar = [col for col in df.columns if col not in columnas_interseccion]
    df.drop(columns=columnas_a_eliminar, inplace=True)
```
